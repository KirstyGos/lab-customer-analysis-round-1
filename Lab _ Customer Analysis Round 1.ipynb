### 9. Perform the data cleaning operations mentioned so far in class

- Delete the column education and the number of open complaints from the dataframe.
- Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.
- Check for duplicate rows in the data and remove if any.
- Filter out the data for customers who have an income of 0 or less.

```python
data.drop(["education", "number_of_open_complaints"], axis = 1)
data.head(2)
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].str.replace(r'%', '')
```

```python
data = data[data["customer_lifetime_value"].notna()]
data.head()
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].astype(float) 
data.head()
```

```python
data['customer_lifetime_value'] = round(data['customer_lifetime_value'],0)
data.head()
```

```python
data['customer_lifetime_value']  = data['customer_lifetime_value'] *100
data.head()
```

```python
data['customer_lifetime_value'] = data['customer_lifetime_value'].astype(int)
data.head()
```
